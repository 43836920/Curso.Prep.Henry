**Question 2:** In how many different states do the customers live in?

```python
# open file and init vars
# evaluate the contents of the file
import pandas as pd

# 1) Lee el CSV
df = pd.read_csv("./sample_data/customers.csv")

# 2) Normaliza el campo State (quita espacios y unifica mayúsculas)
df["State_clean"] = df["State"].astype(str).str.strip().str.upper()

# 3) Cuenta estados distintos
num_estados = df["State_clean"].nunique()

print("Estados diferentes:", num_estados)
```

**Question 3** What is the state with most customers?

```python
# open file and init vars
# evaluate the contents of the file
import pandas as pd

# 1) Lee el CSV
df = pd.read_csv("./sample_data/customers.csv")

# 2) Normaliza el estado
df["State_clean"] = df["State"].astype(str).str.strip().str.upper()

# 3) Para cada CustomerID, elegimos su "estado predominante"
#    (el que más veces aparece para ese cliente). En empate, tomamos el primero alfabéticamente.
state_by_customer = (
    df.groupby("CustomerID")["State_clean"]
      .agg(lambda s: s.value_counts().sort_index().idxmax())
)

# 4) Contamos cuántos clientes viven en cada estado y tomamos el mayor
counts = state_by_customer.value_counts()
estado_mayoria = counts.idxmax()
num_clientes = counts.max()

print("Estado con más clientes:", estado_mayoria)
print("Clientes en ese estado:", num_clientes)
```

**Question 4** What is the state with the least customers?

```python
# open file and init vars
# evaluate the contents of the file
import pandas as pd

# 1) Lee el CSV
df = pd.read_csv("./sample_data/customers.csv")

# 2) Normaliza el estado (evita que "ca", "CA ", etc. cuenten distinto)
df["State_clean"] = df["State"].astype(str).str.strip().str.upper()

# 3) Estado "predominante" por cliente (CustomerID)
#    - Elegimos el estado que más aparece para ese cliente
#    - Si hay empate, se elige el primero alfabéticamente (sort_index)
state_by_customer = (
    df.groupby("CustomerID")["State_clean"]
      .agg(lambda s: s.value_counts().sort_index().idxmax())
)

# 4) Conteo de clientes por estado (ya sin duplicar clientes)
counts = state_by_customer.value_counts()

# 5) Mínimo y estados que tienen ese mínimo (puede haber empate)
min_clientes = counts.min()
estados_minimos = counts[counts == min_clientes].sort_index()

print("Mínimo # de clientes en un estado:", min_clientes)
print("Estado(s) con menos clientes:")
print(estados_minimos)
```

**Question 5:** What is the most common last name?

```python
# open file and init vars
# evaluate the contents of the file
import pandas as pd

# 1) Lee el CSV
df = pd.read_csv("./sample_data/customers.csv")

# 2) Limpia el apellido (quita espacios)
df["Last_clean"] = df["LastName"].astype(str).str.strip()

# 3) Si un cliente aparece más de una vez, elegimos su apellido "predominante"
#    (el que más se repite para ese CustomerID). Si hay empate, toma el primero alfabéticamente.
last_by_customer = (
    df.groupby("CustomerID")["Last_clean"]
      .agg(lambda s: s.value_counts().sort_index().idxmax())
)

# 4) Cuenta apellidos por cliente único y obtiene el/los más comunes (maneja empates)
counts = last_by_customer.value_counts()
max_n = counts.max()
apellidos_top = counts[counts == max_n].sort_index()

print("Apellido(s) más común(es) por cliente único:")
print(apellidos_top)
```

```python
# open and evaluate orders.csv
# Question #4: What is the number of orders placed in October 2021?
import pandas as pd

# 1) Lee el CSV
df = pd.read_csv("./sample_data/Order.csv")  # si tu archivo se llama orders.csv, cambia el nombre aquí

# 2) Convierte Date a datetime (robusto)
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# 3) Filtra octubre 2021 (inclusive)
mask = (df["Date"] >= "2021-10-01") & (df["Date"] < "2021-11-01")
oct_2021 = df.loc[mask]

# 4) Cuenta pedidos únicos
num_pedidos = oct_2021["OrderID"].nunique()

print("Pedidos únicos en octubre 2021:", num_pedidos)
```

```python
# open and evaluate orders.csv
# Question #5: Which customer spent the most amount of money in 2021?
import pandas as pd

# Ajusta estos nombres si en tu Colab se llaman distinto
orders_file = "./sample_data/Order.csv"
customers_file = "./sample_data/customers.csv"

orders = pd.read_csv(orders_file)
customers = pd.read_csv(customers_file)

# 2) Parsear fecha y filtrar 2021
orders["Date"] = pd.to_datetime(orders["Date"], errors="coerce")
orders_2021 = orders[orders["Date"].dt.year == 2021].copy()

# 3) No duplicar pedidos: 1 fila por OrderID (con su total)
orders_2021_unique = (
    orders_2021[["OrderID", "CustomerID", "OrderTotal"]]
    .drop_duplicates(subset=["OrderID"])
)

# 4) Sumar gasto por cliente
spend_by_customer = (
    orders_2021_unique.groupby("CustomerID", as_index=False)["OrderTotal"]
    .sum()
    .rename(columns={"OrderTotal": "TotalSpent2021"})
)

# 5) Obtener el máximo (maneja empates)
max_spend = spend_by_customer["TotalSpent2021"].max()
top = spend_by_customer[spend_by_customer["TotalSpent2021"] == max_spend].copy()

# 6) Unir con customers para traer nombre (ajusta columnas si difieren)
#    Asumo que customers tiene al menos: CustomerID, FirstName, LastName
result = top.merge(customers, on="CustomerID", how="left")

# 7) Mostrar
result["TotalSpent2021"] = result["TotalSpent2021"].round(2)
print("Cliente(s) que más gastó(aron) en 2021:")
display(result[["CustomerID", "FirstName", "LastName", "TotalSpent2021"]])
```

```python
# open and evaluate orders.csv
# Question #6: Historically, what is the best month for sales?
import pandas as pd

# 2) Lee el CSV
df = pd.read_csv("./sample_data/Order.csv")

# 3) Convierte Date a datetime
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# 4) Dedupe por pedido (1 fila por OrderID) para no duplicar ventas
orders_unique = df[["OrderID", "Date", "OrderTotal"]].drop_duplicates(subset=["OrderID"]).copy()

# 5) Extrae el mes y suma ventas por mes (histórico)
orders_unique["Month"] = orders_unique["Date"].dt.month
ventas_por_mes = orders_unique.groupby("Month")["OrderTotal"].sum().sort_values(ascending=False)

# 6) Mejor mes
mejor_mes_num = int(ventas_por_mes.index[0])
mejor_mes_total = float(ventas_por_mes.iloc[0])

# 7) Nombre del mes (en español)
meses = {
    1:"Enero", 2:"Febrero", 3:"Marzo", 4:"Abril", 5:"Mayo", 6:"Junio",
    7:"Julio", 8:"Agosto", 9:"Septiembre", 10:"Octubre", 11:"Noviembre", 12:"Diciembre"
}
print(f"Mejor mes histórico: {meses[mejor_mes_num]} (Mes {mejor_mes_num})")
print(f"Ventas totales: {mejor_mes_total:,.2f}")

print("\nRanking de ventas por mes (desc):")
display(ventas_por_mes.to_frame("TotalSales"))
```
